```python
#Calculo de Media y Mediana por Entidad Federativa
db_total = db[db['Causa'] == 'Total']

columns_of_interest = ['Entidad federativa'] + [f'{year}_T' for year in range(2010, 2023)]
db_filtered = db_total[columns_of_interest].copy()

numeric_columns = [f'{year}_T' for year in range(2010, 2023)]
db_filtered_numeric = db_filtered[numeric_columns]

db_filtered['Media'] = db_filtered_numeric.mean(axis=1)
db_filtered['Mediana'] = db_filtered_numeric.median(axis=1)

result = db_filtered[['Entidad federativa', 'Media', 'Mediana']]
print(result)
```

Entidades como Jalisco, Guanajuato, y México muestran medias relativamente altas, lo que sugiere una mayor cantidad promedio de suicidios en comparación con otras entidades.
La diferencia entre la media y la mediana puede indicar la presencia de valores atípicos en algunos estados. Por ejemplo, Jalisco y Guanajuato tienen una diferencia notable, lo que podría sugerir la influencia de algunos años con tasas de suicidio más altas.
Baja California Sur, Colima, y Tlaxcala tienen medias y medianas relativamente bajas, lo que podría sugerir una menor incidencia de suicidios en estas regiones.
Algunas entidades muestran una mayor consistencia entre la media y la mediana, indicando una distribución más uniforme de los datos a lo largo del período.

```python
#Calculo el rango, percentiles y rango intercuartílico
db_total = db[db['Causa'] == 'Total']
columns_of_interest = ['Entidad federativa'] + [f'{year}_T' for year in range(2010, 2023)]
db_filtered = db_total[columns_of_interest].copy()

numeric_columns = [f'{year}_T' for year in range(2010, 2023)]
db_filtered_numeric = db_filtered[numeric_columns]

db_filtered['Rango'] = db_filtered_numeric.max(axis=1) - db_filtered_numeric.min(axis=1)
db_filtered['Percentil_25'] = np.percentile(db_filtered_numeric, 25, axis=1)
db_filtered['Percentil_75'] = np.percentile(db_filtered_numeric, 75, axis=1)
db_filtered['Rango_intercuartilico'] = db_filtered['Percentil_75'] - db_filtered['Percentil_25']

result = db_filtered[['Entidad federativa', 'Rango', 'Percentil_25', 'Percentil_75', 'Rango_intercuartilico']]
print(result)
```

Se observan notables variaciones en los índices de suicidio a lo largo del país. Aguascalientes presenta el menor rango y el rango intercuartílico más estrecho, sugiriendo cierta estabilidad en las tasas de suicidio. Por otro lado, Chihuahua y México muestran los rangos más amplios, indicando una mayor variabilidad en las tasas de suicidio a lo largo del tiempo.

 Chihuahua, Ciudad de México y Jalisco destacan con valores elevados en ambos percentiles, señalando una mayor dispersión y posiblemente condiciones socioeconómicas diversas que influyen en las tasas de suicidio.

 El rango intercuartílico en entidades como Jalisco, Guanajuato y México exhibe valores notoriamente altos, indicando una variabilidad sustancial en las tasas de suicidio entre distintos periodos. En contraste, Tlaxcala muestra el rango intercuartílico más bajo, sugiriendo una mayor consistencia temporal en las tasas de suicidio.

# **Visualización de datos: Distribuciones**

```python
#Historigrama de Suicidios totales por año
df_filtered = df[['2010_T', '2011_T', '2012_T', '2013_T', '2014_T', '2015_T', '2016_T', '2017_T', '2018_T', '2019_T', '2020_T', '2021_T', '2022_T']]

total_suicides_per_year = df_filtered.sum()

plt.figure(figsize=(10, 6))
plt.bar(total_suicides_per_year.index, total_suicides_per_year.values, color='skyblue')
plt.xlabel('Año')
plt.ylabel('Frecuencia de Suicidios')
plt.title('Histograma de Suicidios Totales por Año (2010-2022)')
plt.show()
```

En el histograma presentado se observa de manera gráfica la frecuencia total de suicidios por año,la cual ha aumentado gradualmente desde 2010, iniciando con una frecuencia anual de 10000 incidencias y llegando al año 2022 con un total de 16000, tuvo un primer crecimiento en 2014 sumando alrededor de 1500 suicidios más de los presentados en 2012, el segundo crecimiento acelerado se muestra en 2019 donde se sumaron al rededor de 2000 suicidios más por año para finalmente disminuir en 2022

```python
#Total de Suicidios de hombres y mujeres por año
df_gender = df[['Entidad federativa', 'Causa'] + [col for col in df.columns if col.endswith(('H', 'M'))]]

df_gender_filtered = df_gender[df_gender['Causa'] != 'Total']
df_grouped = df_gender_filtered.groupby('Causa').sum().transpose()

years = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

labels = [f'{year}_H' for year in years] + [f'{year}_M' for year in years]

df_grouped.plot(kind='bar', figsize=(12, 6))
plt.title('Total de Suicidios de Hombres y Mujeres por Año')
plt.xlabel('Año_Género')
plt.ylabel('Cantidad de Suicidios')
plt.xticks(range(len(labels)), labels, rotation=45, ha='right')
plt.legend(title='Causa')

plt.show()
```

```python
#Boxplot de Causas de Suicidio en México en el periodo 2010-2022
df_sum = db.groupby('Causa').sum().drop('Total')

df_sum_transposed = df_sum.transpose()

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_sum_transposed, orient="h", palette="Set3")
plt.title('Boxplot de Causas de Suicidio en México (2010-2022)')
plt.xlabel('Número de Suicidios')
plt.ylabel('Causas')
plt.show()
```

```python
#Distribuición de Suicidios por Genero en México
df_total = df[df['Causa'] == 'Total']

total_hombres = df_total.iloc[:, 2::3].sum(axis=1)
total_mujeres = df_total.iloc[:, 3::3].sum(axis=1)

df_genero = pd.DataFrame({'Hombres': total_hombres, 'Mujeres': total_mujeres})

total_por_genero = df_genero.sum()

plt.figure(figsize=(8, 8))
plt.pie(total_por_genero, labels=total_por_genero.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de suicidios por género')
plt.show()
```

```python
#Distribución de suicidios en hombres y mujeres por causa
cols_hombres = ['Entidad federativa', 'Causa'] + [str(year) + '_H' for year in range(2010, 2023)]
cols_mujeres = ['Entidad federativa', 'Causa'] + [str(year) + '_M' for year in range(2010, 2023)]

df_hombres = df[cols_hombres]
df_mujeres = df[cols_mujeres]

df_hombres_agrupado = df_hombres.groupby('Causa', as_index=False).sum(numeric_only=True)
df_mujeres_agrupado = df_mujeres.groupby('Causa', as_index=False).sum(numeric_only=True)

causas_interes = ['Ahorcamiento estrangulamiento o sofocación ', 'Disparo de arma de fuego', 'Envenenamiento', 'Otras causas']

df_hombres_causas_interes = df_hombres_agrupado[df_hombres_agrupado['Causa'].isin(causas_interes)].copy()
df_hombres_causas_interes['Total'] = df_hombres_causas_interes.iloc[:, 2:].sum(axis=1)

df_mujeres_causas_interes = df_mujeres_agrupado[df_mujeres_agrupado['Causa'].isin(causas_interes)].copy()
df_mujeres_causas_interes['Total'] = df_mujeres_causas_interes.iloc[:, 2:].sum(axis=1)

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.pie(df_hombres_causas_interes['Total'], labels=df_hombres_causas_interes['Causa'], autopct='%1.1f%%', startangle=140)
plt.title('Distribución de suicidios en hombres por causa (2010-2022)')

plt.subplot(1, 2, 2)
plt.pie(df_mujeres_causas_interes['Total'], labels=df_mujeres_causas_interes['Causa'], autopct='%1.1f%%', startangle=140)
plt.title('Distribución de suicidios en mujeres por causa (2010-2022)')

plt.show()
```

Las dos gráficas de pastel representan la distribución de las causas de suicidio en hombres y mujeres durante el periodo de 2010 a 2022. En el caso de los hombres, la mayoría de los suicidios se atribuyen a ahorcamiento, estrangulamiento o sofocación, con un impresionante 82.5% de los casos. Le siguen el suicidio por disparo de arma de fuego, que representa el 8.8%, envenenamiento con un 5.2%, y otras causas con un 3.4%. Estos porcentajes ilustran claramente la predominancia del método de ahorcamiento entre los hombres durante este periodo.

Por otro lado, en el caso de las mujeres, el 75.9% de los suicidios se relacionan con ahorcamiento, estrangulamiento o sofocación, evidenciando también una alta incidencia de esta causa. Sin embargo, las proporciones varían significativamente en comparación con los hombres. Las mujeres muestran un 3.3% de casos de suicidio por disparo de arma de fuego, un 16.5% por envenenamiento y un 4.2% por otras causas. La distribución porcentual destaca las diferencias en los métodos utilizados por mujeres en comparación con los hombres.

Al comparar ambas gráficas, se destaca la marcada disparidad en las causas de suicidio entre hombres y mujeres. Mientras que el ahorcamiento es la principal causa para ambos grupos, la magnitud de esta tendencia es mucho más pronunciada entre los hombres. Además, la proporción de suicidios por disparo de arma de fuego, envenenamiento y otras causas es considerablemente mayor en hombres en comparación con mujeres, indicando diferencias significativas en los métodos preferidos para ambos géneros durante el período analizado.
